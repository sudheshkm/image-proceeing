```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Load an image
image = cv2.imread("/content/drive/MyDrive/Workshop/cluster2.jpeg")

# Convert the image to a 2D array of pixels (reshape it)
pixels = image.reshape((-1, 3))

# Convert the data type of the pixels to floating-point
pixels = np.float32(pixels)

# Define the number of clusters (K)
K = 3

# Define criteria for the K-means algorithm
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)

# Apply K-means clustering
_, labels, centers = cv2.kmeans(pixels, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Convert the centers of clusters to 8-bit integers
centers = np.uint8(centers)

# Map the labels to the centers
segmented_image = centers[labels.flatten()]

# Reshape the segmented image to its original shape
segmented_image = segmented_image.reshape(image.shape)

# Display the original and segmented images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title("Original Image")
plt.subplot(122), plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB)), plt.title("Segmented Image")
plt.show()
```

 In this case, EPS stands for epsilon, which represents the threshold for the relative change.

 100: This is the maximum number of iterations allowed for the algorithm. If the algorithm does not converge within these 100 iterations, it will terminate.

0.2: This is the epsilon threshold for the cv2.TERM_CRITERIA_EPS condition. It defines the acceptable relative change in the objective function between consecutive iterations that indicates convergence
